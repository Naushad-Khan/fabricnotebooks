# Parameters

```python
workspaces = ["af62e809-b9c7-4d02-8633-869d4718823d"]
targetCapacityId = "631FE32D-FE8B-4E9B-B6A5-61C35DD79FA7"
```

# Script

```python
client = fabric.FabricRestClient()

for workspaceId in workspaces:
      
    payload = {
        'capacityMigrationAssignments': [{
            'targetCapacityObjectId': targetCapacityId
            , 'workspacesToAssign': workspaces
        }]
    }        

    response = client.post(f"/v1.0/myorg/admin/capacities/AssignWorkspaces", json= payload)

    response
```
