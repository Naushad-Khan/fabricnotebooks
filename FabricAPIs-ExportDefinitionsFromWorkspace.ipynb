```python
# Workspace to get all item definitions

workspaceId = "c511343c-b6bc-4b6e-b780-12a1b634e60e"

# Folder in the lakehouse to store the files
baseDir = f"/lakehouse/default/Files/definitions/{workspaceId}" 
```

```python
client = fabric.FabricRestClient()

items = fabric.list_items(workspace=workspaceId)

for index, item in items.iterrows():

    try:
        response = client.post(f"/v1/workspaces/{workspaceId}/items/{item.Id}/getDefinition", lro_wait=True)

        jsonParts = response.json()['definition']['parts']

        for part in jsonParts:

            print(f"Downloaded part '{part['path']}'") 

            if (baseDir != ""):            
                payload = part['payload']

                payloadBytes = base64.b64decode(payload)

                filePath = os.path.join(baseDir, f"{item.Id}/{part['path']}")

                print(f"Saving to lake: '{filePath}'")

                os.makedirs(os.path.dirname(filePath), exist_ok=True)

                with open(filePath, "wb") as file:
                    file.write(payloadBytes)
    except Exception as ex:
        print(f"Error downloading item '{item.Id}'. Error: '{str(ex)}'")
```
